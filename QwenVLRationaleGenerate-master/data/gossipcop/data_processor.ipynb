```python

import json
import os

import pandas as pd



label_dict = {
    'real':1,
    'fake':0
}

def read_origin_data():
    with open('gossipcop_v3_origin.json','r',encoding='utf8') as f:
        data = json.load(f)
        df = pd.DataFrame.from_dict(data,orient='index')
        df = df[['id','text','label','publish_date']]
        df['label'].apply(lambda x: label_dict[x])
        return df

df = read_origin_data()

df[:10]
        
```

```python


def get_image_set():
    # 获取 images 目录下所有图像文件的前缀
    return {image_f.split('_')[0] for image_f in os.listdir('images')}

def filter_no_image_data(df):
    # 根据 id 过滤 DataFrame，只保留存在于 image_set 中的行
    image_set = get_image_set()
    return df[df['id'].isin(image_set)]

def filter_data(df):
    # 过滤掉没有图像的行
    df = filter_no_image_data(df)
    # 删除没有 publish_date 的行
    # df = df.dropna(subset=['publish_date'])
    # 将 publish_date 列转换为 datetime 格式
    # df['publish_date'] = pd.to_datetime(df['publish_date'], unit='s')
    # # 格式化为字符串
    # df['publish_date'] = df['publish_date'].dt.strftime('%Y-%m-%d %H:%M:%S')
    return df
 
print(f"before filter all : {len(df)} real : {(df['label']=='real').sum()} fake: {(df['label'] == 'fake').sum()}")
df = filter_data(df)
print(f"after filter all : {len(df)} real : {(df['label']=='real').sum()} fake : {(df['label'] == 'fake').sum()}")
```
