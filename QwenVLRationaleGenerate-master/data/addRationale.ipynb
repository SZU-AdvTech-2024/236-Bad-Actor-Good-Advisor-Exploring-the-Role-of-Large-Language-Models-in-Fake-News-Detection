```python
import  pandas as pd

rationale_name = 'multi'
rationale_df = pd.read_csv(f'/home/lyq/DataSet/FakeNews/gossipcop/{rationale_name}.csv')
df = pd.read_csv('/home/lyq/DataSet/FakeNews/gossipcop/gossipcop_llm_rationales.csv')
rationale_df.shape[0],df.shape[0]
```

```python
from data_loader import label_str2int_dict

rationale_df.rename(columns={'authenticity':f'{rationale_name}_pred','reason':f'{rationale_name}_rationale'},inplace=True)
rationale_df[f'{rationale_name}_pred'] = rationale_df[f'{rationale_name}_pred'].apply(lambda x: label_str2int_dict[x])
rationale_df
```

```python
df = df.merge(rationale_df,how='left',on='source_id')

df.columns
```

```python
df[f'{rationale_name}_acc'] = (df[f'{rationale_name}_pred'] == df['label']).astype(int)


df
```

```python
rationale_names = [col_name.split('_')[0] for col_name in df.columns if col_name.endswith('_rationale')]
rationale_names
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score


def cal_rationale_metrics(y_pred,y_true):
    return {
        'acc' : accuracy_score(y_true, y_pred),
        'recall': recall_score(y_true, y_pred,average='macro'),
        'recall_real': recall_score(y_true, y_pred,average=None,labels=[0]),
        'recall_fake': recall_score(y_true, y_pred,average=None,labels=[1]),
        'precision': precision_score(y_true, y_pred,average='macro'),
        'precision_real': precision_score(y_true, y_pred,average=None,labels=[0]),
        'precision_fake': precision_score(y_true, y_pred,average=None,labels=[1]),
    }


for rationale_name in rationale_names:
    print(rationale_name,cal_rationale_metrics(df[f'{rationale_name}_pred'].to_numpy(),df['label'].to_numpy()))
```
